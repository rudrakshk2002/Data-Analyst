# plot 

```python
df.columns
```

```python
ax=sns.countplot(x= 'Gender', data= df)
plt.title('Gender Plot')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
plt.figure(figsize=(16,9))
sales= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
sns.barplot(x='Gender', y='Amount', data= sales ,label=['Gender','Amount'])
plt.title('Gender & Amount Plot')
plt.legend()
plt.show()
```

# AGE

```python
ax1=sns.countplot(x= 'Age Group', data= df)
```

```python
plt.figure(figsize=(16,9))
ax1=sns.countplot(x= 'Age Group', hue='Gender', data= df)
plt.title('Gender & Age groups Plot')
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize=(16,9))

sales= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sales)
plt.title('Age Group & Amount Plot')
plt.show()
```

# STATE

```python
plt.figure(figsize=(15,5))
ax=sns.countplot(x= 'State', data= df)
plt.title('state Plot' , fontsize=40)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
plt.figure(figsize=(15,5))

sales1= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by=['Orders'], ascending=False).head(10)
sns.barplot(x='State', y='Orders', data= sales1)
plt.title('state & order Plot', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,5))

sales1= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False).head(10)
sns.barplot(x='State', y='Amount', data= sales1)
plt.title('state & amount Plot', fontsize=20)
plt.show()
```
