# State Count

# Male & Female

```python
# Melt the DataFrame to have a 'Gender' column and a 'Count' column
df_melted = df.melt(value_vars=['Male', 'Female'], var_name='Gender', value_name='Count')

# Create a countplot
ax = sns.barplot(x='Gender', y='Count', data=df_melted, estimator=sum)
plt.title('Count of Male and Female')

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container)

# Show the plot
plt.show()
```

```python
 #Melt the DataFrame to have a 'Gender' column and a 'Count' column
df_melted1 = df.melt(value_vars=['Male_Workers', 'Female_Workers'], var_name='Gender worker', value_name='Count')

# Create a countplot
ax = sns.barplot(x='Gender worker', y='Count', data=df_melted1, estimator=sum)
plt.title('Count of Male and Female Workers')

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container)

# Show the plot
plt.show()
```

# Education Level

```python
# Melt the DataFrame to long format
df_melted = df.melt(id_vars=['State_name'], value_vars=['Secondary_Education', 'Higher_Education', 'Graduate_Education'], var_name='Education_Level', value_name='Count')

# Plot
plt.figure(figsize=(10, 6))
education_plot = df_melted.pivot_table(index='State_name', columns='Education_Level', values='Count')
education_plot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Education Levels by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.legend(title='Education Level')
plt.show()
```

# Age Groups

```python
# Melt the DataFrame to long format
df_melted = df.melt(id_vars=['State_name'], value_vars=['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50'], var_name='Age_Group', value_name='Count')

# Plot
plt.figure(figsize=(10, 6))
age_plot = df_melted.pivot_table(index='State_name', columns='Age_Group', values='Count')
age_plot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Age Group Distribution by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.legend(title='Age Group')
plt.show()
```

# Population Distribution by State

```python
# Assuming df is your DataFrame
population_by_state = df.groupby('State_name')['Population'].sum().reset_index()

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='Population', y='State_name', data=population_by_state)
plt.title('Total Population by State')
plt.xlabel('Population')
plt.ylabel('State')
plt.show()
```

# Education Levels by Gender

```python
education_by_gender = df[['Male', 'Female', 'Secondary_Education', 'Higher_Education', 'Graduate_Education']]

# Summing up the education levels
education_sum = education_by_gender[['Secondary_Education', 'Higher_Education', 'Graduate_Education']].sum()

# Plotting
plt.figure(figsize=(10, 6))
education_sum.plot(kind='bar', color=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Education Levels Across the Population')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

# Correlation Matrix

```python
# Select columns of interest
columns_of_interest = ['Population', 'Male', 'Female', 'Literate', 'Workers', 
                       'Secondary_Education', 'Higher_Education', 'Graduate_Education']

# Calculate correlation
correlation_matrix = df[columns_of_interest].corr()

# Plotting
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Key Variables')
plt.show()
```

# Age Group Distributio

```python
age_group_distribution = df[['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50']].sum()

# Plotting
plt.figure(figsize=(8, 6))
age_group_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Age Group Distribution')
plt.show()
```
